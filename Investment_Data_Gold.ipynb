```python
df_silver = spark.read.table("workspace.default.silver_gold_18k")
df_forecast = spark.read.table("workspace.default.forecast_monthly_prediction")
```

```python
from pyspark.sql.functions import to_date

df_forecast = df_forecast.withColumn("ds", to_date("ds"))
```

```python
df_gold = df_silver.join(df_forecast, df_forecast.ds == df_silver.date, how="outer")
```

```python
df_gold.write.format("delta").mode("overwrite").save("/Volumes/workspace/default/invest/gold_gold_18k")
df_gold.write.mode("overwrite").saveAsTable("gold_gold_18k")
```
